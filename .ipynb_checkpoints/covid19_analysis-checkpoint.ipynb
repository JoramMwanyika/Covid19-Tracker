# COVID-19 Global Data Tracker

This notebook analyzes global COVID-19 trends including cases, deaths, and vaccinations across different countries and time periods.

## Data Source
We use the Our World in Data (OWID) COVID-19 dataset.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from datetime import datetime

# Set plot style
plt.style.use('seaborn')
sns.set_palette('husl')
```

## 1. Data Collection and Loading
Download and load the OWID COVID-19 dataset

```python
# Download the latest dataset
import requests

url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
response = requests.get(url)

# Save the file
with open('owid-covid-data.csv', 'wb') as f:
    f.write(response.content)

# Load the dataset
df = pd.read_csv('owid-covid-data.csv')
print(f"Dataset shape: {df.shape}")
df.head()
```
